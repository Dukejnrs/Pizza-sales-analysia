```python
# Pizza Place Sales Analysis
# Task 3 for Data Science Course
# Analyzes pizza sales data to answer required questions and explore trends.

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)

# === Load and Merge Data ===
# File path in Google Drive
data_path = '/content/drive/My Drive/Pizza_analysis/'

# Load CSV files with latin1 encoding
try:
    orders = pd.read_csv(data_path + 'orders.csv', encoding='latin1')
    order_details = pd.read_csv(data_path + 'order_details.csv', encoding='latin1')
    pizzas = pd.read_csv(data_path + 'pizzas.csv', encoding='latin1')
    pizza_types = pd.read_csv(data_path + 'pizza_types.csv', encoding='latin1')
except FileNotFoundError:
    print("Error: CSV files not found at", data_path)
    print("Check folder: /content/drive/My Drive/Pizza_analysis/")
    raise
except Exception as e:
    print(f"Error loading files: {e}")
    raise

# Merge dataframes
df = pd.merge(orders, order_details, on='order_id', how='left')
df = pd.merge(df, pizzas, on='pizza_id', how='left')
df = pd.merge(df, pizza_types, on='pizza_type_id', how='left')

# Display preview
print("Merged Dataframe Preview:")
display(df.head())
print("\nColumns:", df.columns.tolist())

# === Data Cleaning ===
# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Convert date and time
try:
    df['date'] = pd.to_datetime(df['date'])
    df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S').dt.time
except KeyError:
    print("Error: 'date' or 'time' column not found. Check data dictionary.")
    raise

# === Recommended Analysis ===

# 1. Total Revenue/Sales
try:
    total_revenue = (df['price'] * df['quantity']).sum()
    print(f"\nTotal Revenue: ${total_revenue:.2f}")
except KeyError:
    print("Error: 'price' or 'quantity' column not found.")

# 2. Total Quantity Sold
total_quantity = df['quantity'].sum()
print(f"Total Quantity Sold: {total_quantity}")

# 3. Total Orders
total_orders = df['order_id'].nunique()
print(f"Total Orders: {total_orders}")

# 4. Number of Pizza Types
num_pizza_types = df['pizza_type_id'].nunique()
print(f"Number of Pizza Types: {num_pizza_types}")

# 5. Average Price of Pizzas
avg_price = df['price'].mean()
print(f"Average Pizza Price: ${avg_price:.2f}")

# 6. Peak Hours of Sales
df['hour'] = pd.to_datetime(df['time'], format='%H:%M:%S').dt.hour
sales_by_hour = df.groupby('hour')['quantity'].sum()
print("\nSales by Hour:")
print(sales_by_hour)

# Plot peak hours
plt.figure(figsize=(10, 6))
sales_by_hour.plot(kind='bar')
plt.title('Sales by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Quantity Sold')
plt.savefig('/content/drive/My Drive/Pizza_analysis/sales_by_hour.png')
plt.show()

# 7. Total Sales by Day of Week
df['day_of_week'] = df['date'].dt.day_name()
sales_by_day = df.groupby('day_of_week')['price'].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)
print("\nSales by Day of Week:")
print(sales_by_day)

# Plot sales by day
plt.figure(figsize=(10, 6))
sales_by_day.plot(kind='bar')
plt.title('Sales by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Total Sales ($)')
plt.savefig('/content/drive/My Drive/Pizza_analysis/sales_by_day.png')
plt.show()

# 8. Top 5 Bestselling Pizzas
top_pizzas = df.groupby('name')['quantity'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Bestselling Pizzas:")
print(top_pizzas)

# 9. Sales by Month
df['month'] = df['date'].dt.month_name()
sales_by_month = df.groupby('month')['price'].sum()
print("\nSales by Month:")
print(sales_by_month)

# Plot sales by month
plt.figure(figsize=(10, 6))
sales_by_month.plot(kind='bar')
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.savefig('/content/drive/My Drive/Pizza_analysis/sales_by_month.png')
plt.show()

# 10. Pizza Types Not Doing Well
bottom_pizzas = df.groupby('name')['quantity'].sum().sort_values().head(5)
print("\nBottom 5 Pizza Types (Least Sold):")
print(bottom_pizzas)

# === Additional Exploration ===
# Average Order Value
avg_order_value = df.groupby('order_id')['price'].sum().mean()
print(f"\nAverage Order Value: ${avg_order_value:.2f}")

# Sales by Pizza Size
sales_by_size = df.groupby('size')['quantity'].sum()
print("\nSales by Pizza Size:")
print(sales_by_size)

# Average Pizzas per Order
avg_pizzas_per_order = df.groupby('order_id')['quantity'].sum().mean()
print(f"Average Pizzas per Order: {avg_pizzas_per_order:.2f}")

# Save merged dataframe
df.to_csv('/content/drive/My Drive/Pizza_analysis/merged_pizza_data.csv', index=False)
print("Saved: merged_pizza_data.csv")
```
